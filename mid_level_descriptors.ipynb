Dies ist ein kleiner Ausflug in die Linguistik. Hierbei handelt es sich um eine Spracherkennung, welche  automatisch die Rede transkribbiert hat. 'ASR' steht für 'Automatic Speech Recognition'. Falls ich das richtig verstanden habe, hat die automatische Spracherkennung mit der Hilfe von Google funktioniert.
Diese Funktion hat jedoch zu großen Problemen geführt und ist immer wieder zu willkürlichen Zeitpunkten abgestürzt. 

```python
import asr as speechrec
asr = speechrec.ASR('deu')
df_asr = asr.extract_from_index(df_mld_seqs, tmp='./tmp', margin=0.2)
```

```python
df_asr.head()
```

Ebenfalls ein von Uwe Reichel programmiertes Package. Das Package dient hierbei zur Textanalyse und soll bspw. automatisch  erkennen, wie viele Adjektive, Adverben, Konjunktionen, Funktionswörter, etc. in der Rede vorkamen.
Vorsicht: Hier kann es schnell zu einem ZeroDivisionError kommen, da bei dieser linguistischen Analyse eher ganze Sätze erwartet werden. Hier bei gibt es jedoch immer Mal wieder kurze Satzfragmente oder sogar einzelne Wörter, welche zu diesem Fehler führen können. Dementsprechend wurde dieser Fehler hier 'gefangen', indem die 'except'Regelung hier mit eingefügt wurde.
Der Fehler könnte erscheinen, weil irgendetwas gezählt wird, was eigentlich null ist und da man durch null nicht teilen darf.

Auch hier in der Schleife wird mit dem t_all - Befehl der ganze Text aneinander gehangen.
Dies sieht man dann auch im späteren Verlauf: Dort kann der ganze, zusammenhängende Text nur noch mit 't_all' aufgerufen werden.

```python
fex = tf.PsyText(language='deu')
ling_df = pd.DataFrame()
t_all = ''
for t in df_asr.asr:
    print(t)
    t_all = t_all+' '+t
    try:
        feat = fex.extract_from_string(t)
    except ZeroDivisionError:
        feat = pd.DataFrame()
    ling_df = ling_df.append(feat)
```
